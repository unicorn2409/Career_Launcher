# 1.5

```python
# Calculate average price of last N days......
def avg_price(N):
    return (data2['Average Price'].tail(N).sum())/N
print("Average prices for last N days are as follows:")
print("Last 1 week",avg_price(5))
print("Last 2 weeks",avg_price(10))
print("Last 1 month",avg_price(20))
print("Last 3 months",avg_price(60))
print("Last 6 months",avg_price(120))
print("Last 1 year",avg_price(240))

# Calculate profit/loss percentage over last N days.......
print("Profit/Loss % for N days are as follows:")
def prof_loss(N):
    difference = (data2['Close Price'].tail(N).iloc[N-1] - data2['Close Price'].tail(N).iloc[0])
    if difference < 0 :
        loss = -(difference)
        loss_percen = (loss/data2['Close Price'].tail(N).iloc[N-1])*100
        return loss_percen
    if difference > 0 :
        profit = difference
        profit_percen = (profit/data2['Close Price'].tail(N).iloc[N-1])*100
        return profit_percen
print("Loss/Profit percentage for last N days are as follows:")
print("Last 1 week",prof_loss(5))
print("Last 2 weeks",prof_loss(10))
print("Last 1 month",prof_loss(20))
print("Last 3 months",prof_loss(60))
print("Last 6 months",prof_loss(120))
print("Last 1 year",prof_loss(240))
```

# 1.7

```python
data2["Trend"] = "Slight or No change"
data2
```

```python
if ((data2['Day_Perc_Change'] >= -0.5) & (data2['Day_Perc_Change'] <= 0.5)).all():
    data2['Trend'] = 'Slight or No change'
if ((data2['Day_Perc_Change'] >= 0.5) & (data2['Day_Perc_Change'] <= 1)).all():
    data2['Trend'] = 'Slight positive'
if ((data2['Day_Perc_Change'] <= -0.5) & (data2['Day_Perc_Change'] >= -1)).all():
    data2['Trend'] = 'Slight negative'
if ((data2['Day_Perc_Change'] >= 1) & (data2['Day_Perc_Change'] <= 3)).all():
    data2['Trend'] = 'Positive' 
if ((data2['Day_Perc_Change'] <= -1) & (data2['Day_Perc_Change'] >= -3)).all():
    data2['Trend'] = 'Negative'
if ((data2['Day_Perc_Change'] >= 3) & (data2['Day_Perc_Change'] <= 7)).all():
    data2['Trend'] = 'Among top gainers'
if ((data2['Day_Perc_Change'] <= -3) & (data2['Day_Perc_Change'] >= -7)).all():
    data2['Trend'] = 'Among top losers'
if (data2['Day_Perc_Change'] > 7).all():
    data2['Trend'] = 'Bull run' 
if (data2['Day_Perc_Change'] < -7).all():
    data2['Trend'] = 'Bear drop' 
data2
```
