```python
import pandas as pd
import matplotlib.pyplot as plt

tcs_data = pd.read_csv('D:\\Study\\TY_BTECH\Resume\\Internship\\CL_Data_School\\ML_Project\\tcs_stock_data.csv')
data6 = tcs_data
data6
```

```python
 # Create a necesarry column and put the data into it using mathematical formula........
data6['15_day_Merge_Average'] = data6['Close Price'].rolling(window=15).mean()
data6['30_days_STD'] = data6['Close Price'].rolling(window=20).std()
data6['upper_bollinger_band'] = data6['15_day_Merge_Average'] + (data6['30_days_STD'] * 2)
data6['lower_bollinger_band'] = data6['15_day_Merge_Average'] - (data6['30_days_STD'] * 2)
data6['middle_bollinger_band'] = (data6['15_day_Merge_Average'] + data6['30_days_STD'])/2
data6
```

```python
# drop unusable rows.......
data6 = data6.dropna()
data6
```

```python
# Function for creating 'call' column and put the entries into it based on the values...........
def buy(data6):
    if data6['Close Price'] < data6['lower_bollinger_band']:
        return 'Buy'
    if data6['Close Price'] > data6['lower_bollinger_band'] and data6['Close Price'] < data6['middle_bollinger_band']:
        return 'Hold Buy / Liquidate Short'
    if data6['Close Price'] > data6['middle_bollinger_band'] and data6['Close Price'] < data6['upper_bollinger_band']:
        return 'Hold Short / Liquidate Buy'
    if data6['Close Price'] > data6['upper_bollinger_band']:
        return 'Short'
data6 = data6.assign(Call = data6.apply(buy, axis=1))
data6
```

```python
# Training of classification problem
import numpy as np
from sklearn import preprocessing
# create a object for LabelEncoder()
LE = preprocessing.LabelEncoder()
train_model = data6[['upper_bollinger_band', 'lower_bollinger_band', 'middle_bollinger_band', 'Close Price']]

# Now transformed label are fit into the column 'Call'
Output_Label = LE.fit_transform(data6['Call'])
New_Train_Model = Output_Label.reshape(-1,1)
Rows = np.size(New_Train_Model,0)
Columns = np.size(New_Train_Model,1)
print(Rows,',',Columns)
```

```python
# Suport Vector Machine
import sklearn as sk 
from sklearn import svm
import warnings
warnings.filterwarnings('ignore')

from sklearn.linear_model import LogisticRegression
LR = LogisticRegression(random_state=0, solver='lbfgs', multi_class='ovr').fit(train_model, New_Train_Model.ravel())
LR.predict(train_model) 
print("Logistic Regression:",round(LR.score(train_model, New_Train_Model), 4))

SVM = svm.LinearSVC()
SVM.fit(train_model, New_Train_Model)
SVM.predict(train_model)
print("Support Vector Machines Classifier:", round(SVM.score(train_model,New_Train_Model), 4))

# Random Forest
from sklearn.ensemble import RandomForestClassifier
RF = RandomForestClassifier(n_estimators=100, max_depth=2, random_state=0)  
RF.fit(train_model, New_Train_Model)  
RF.predict(train_model) 
print("Random Forests Classifier: ",round(RF.score(train_model, New_Train_Model), 4))

#neural_network Classifier
from sklearn.neural_network import MLPClassifier
NN = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(5, 2), random_state=1)  
NN.fit(train_model, New_Train_Model)  
NN.predict(train_model)
print("Neural Networks Classifier: ",round(NN.score(train_model, New_Train_Model), 4))
```

```python
import pandas as pd
wipro_data = pd.read_csv('D:\\Study\\TY_BTECH\\Resume\\Internship\\CL_Data_School\\ML_Project\\wipro_stock_data.csv')
data7 = wipro_data
data7
```

```python
# Column Defining......
data7['%chg_op_cl'] = ((data7['Close Price'] - data7['Open Price'])/(data7['Close Price']))*100
data7['%chg lw_hg'] = ((data7['Close Price'] - data7['High Price'])/(data7['Low Price']))*100
data7['%chg 5dymean'] = data7['Close Price'].pct_change().dropna().rolling(5).mean()
data7['%chg 5dystd'] = data7['Close Price'].pct_change().dropna().rolling(5).std()
data7 = data7.dropna()
data7
```

```python
arr = []
val = []
for value in data7['Close Price'].iteritems():
    arr.append(value)
#arr
#len(arr)#.......490
for i in range(0,489):
    if arr[i+1] > arr[i]:
        val.append(1)
    else:
        val.append(-1)
data7['Action'] = pd.DataFrame(val)
data7 = data7.dropna()
data7
```

```python
#training a classifier
from sklearn import preprocessing
# create the Labelencoder object
# le = preprocessing.LabelEncoder()
train_model = data7[['%chg_op_cl','%chg lw_hg','%chg 5dymean','%chg 5dystd']]
New_Train_Model = data7[['Action']]
# trafomed_label = le.fit_transform(tcs_dat[['Call']])
# train_Y = transfomed_label.reshape(-1,1)

import sklearn as sk  
from sklearn.ensemble import RandomForestClassifier
RF = RandomForestClassifier(n_estimators=100, max_depth=2, random_state=0)  
RF.fit(train_model, New_Train_Model)  
RF.predict(train_model) 
print("Random Forests:", round(RF.score(train_model,New_Train_Model), 4))
```
