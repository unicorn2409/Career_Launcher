```python
gold_without_Nan = data5.dropna()
gold_without_Nan
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
y = np.array(gold_without_Nan["Pred"])
print('array(Pred)-------------------------\n',y)
x = np.array(gold_without_Nan['new'])
print('array(new)--------------------------\n',x)

y = y.reshape(-1,1)#convert into 2-D array
#411-----without 'none' values
x = x.reshape(-1,1)
#411----without 'none' values
print('reshape(Pred)-----------------------\n',y)
print('reshape(new)------------------------\n',x)
```

```python
# sckit-learn implementation
# Model initialization
regression_model = LinearRegression()

# Fit the data(train the model)
regression_model.fit(x, y)

# Predict
y_predicted = regression_model.predict(x)
#print(y_predicted)

# model evaluation(Root_Mean_Squared_error)
rmse = mean_squared_error(y, y_predicted)
r2 = r2_score(y, y_predicted)

# printing values
print('Slope(Alpha):' ,regression_model.coef_)
print('Intercept:', regression_model.intercept_)
print('Root mean squared error: ', rmse)
print('R2 score: ', r2)

# plotting values
# data points
print('\n**Ploting Data Points-----For 411-Labels')
plt.scatter(x, y, s=10)
plt.xlabel('x')
plt.ylabel('y')

# predicted values
plt.plot(x, y_predicted, color='r')
plt.show()
```

```python
import pandas as pd
import statsmodels.api as sm

tcs = pd.read_csv('D:\Study\TY_BTECH\Resume\Internship\CL_Data_School\TCS.NS.csv', parse_dates=True, index_col='Date',)
nifty50 = pd.read_csv('D:\Study\TY_BTECH\Resume\Internship\CL_Data_School\^NSEI.csv', parse_dates=True, index_col='Date')

# joining the closing prices of the two datasets 
monthly_prices = pd.concat([tcs['Close'], nifty50['Close']], axis=1)
monthly_prices.columns = ['TCS', 'NIFTY50']
# check the head of the dataframe
print('\n**First 5---------------::')
print(monthly_prices.head())

# -------------------------monthly returns--------------------
print('\n**Monthly Returns-------::')
monthly_returns = monthly_prices.pct_change(1)
clean_monthly_returns = monthly_returns.dropna(axis=0)  # drop first missing row
print(clean_monthly_returns.head())
```

```python
# split dependent and independent variable
X = clean_monthly_returns['TCS']
y = clean_monthly_returns['NIFTY50']

# Add a constant to the independent value
X1 = sm.add_constant(X)

# make regression model 
model = sm.OLS(y, X1)

# fit model and print results
results = model.fit()
print(results.summary())
```
